# 01 — Raw data EDA (before annotation)

Quick look at the filtered comment pool and the keyword-matched candidate pool. Run this *before* annotation to motivate the lexicon design and split strategy.

```python
import sys, pathlib
sys.path.append(str(pathlib.Path.cwd().parent))
import pandas as pd, matplotlib.pyplot as plt
from src.paths import CANDIDATES_HATE, CANDIDATES_RANDOM, COMMENTS_FILTERED
```

## Filtered comment pool

```python
df = pd.read_csv(COMMENTS_FILTERED)
print(f'Comments: {len(df):,}')
print(f'Videos:   {df["video_id"].nunique()}')
print(f'Mean words: {df["n_words"].mean():.1f}')
print(f'With Kyrgyz-specific letters: {df["has_kyrgyz_letters"].mean()*100:.1f}%')
df.head(5)
```

```python
df['n_words'].hist(bins=60, color='#7eb0ff')
plt.title('Comment length (words)')
plt.xlabel('words'); plt.ylabel('count'); plt.show()
```

## Candidate pools

```python
hate = pd.read_csv(CANDIDATES_HATE)
rand = pd.read_csv(CANDIDATES_RANDOM)
print(f'Keyword candidates: {len(hate)}')
print(f'Random non-matching: {len(rand)}')
cats = hate['matched_categories'].fillna('').str.split('|').explode().str.strip()
cats = cats[cats != ''].value_counts()
cats.plot(kind='barh', color='#9c89b8')
plt.title('Keyword hits by category'); plt.xlabel('comments'); plt.show()
```

```python
for _, row in hate.sample(5, random_state=0).iterrows():
    print(f'[{row["matched_categories"]:20}] {row["text"][:150]}')
    print()
```
