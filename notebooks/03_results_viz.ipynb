# 03 — Interactive results exploration

Run after `python -m src.run_all`. Loads the experiment summary and lets you slice the results table interactively.

```python
import sys, pathlib
sys.path.append(str(pathlib.Path.cwd().parent))
import pandas as pd
from src.paths import SUMMARY, RESULTS, LABELS
```

```python
summary = pd.read_csv(SUMMARY)
summary.sort_values('macro_f1', ascending=False)
```

```python
best = summary.sort_values('macro_f1', ascending=False).iloc[0]['name']
preds = pd.read_csv(RESULTS / best / 'predictions.csv')
errors = preds[preds['gold'] != preds['pred']]
print(f'best model: {best}')
print(f'errors: {len(errors)}/{len(preds)} = {len(errors)/len(preds)*100:.1f}%')
errors.head(10)
```

```python
# False-positive hate predictions (model says hate, gold says non_hate)
fp_hate = errors[(errors['pred']=='hate') & (errors['gold']=='non_hate')]
for _, r in fp_hate.head(10).iterrows():
    print(r['text'][:200]); print()
```

```python
# Missed-hate (model says non_hate, gold says hate)
miss = errors[(errors['pred']=='non_hate') & (errors['gold']=='hate')]
for _, r in miss.head(10).iterrows():
    print(r['text'][:200]); print()
```
